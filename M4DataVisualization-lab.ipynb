# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
import pandas as pd

QUERY = """
SELECT ConvertedComp
FROM master
"""

df_Conv = pd.read_sql_query(QUERY, conn)
df_Conv.head()


plt.hist(df_Conv, bins=10, color='orange', edgecolor='black')
plt.title('Histogram of ConvertedComp')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
group by age
order by age
"""
df_Age= pd.read_sql_query(QUERY,conn)
df_Age.plot(kind='box', figsize = (10,7))
plt.title('Box plot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
group by age
order by age
"""
df_AgeWork= pd.read_sql_query(QUERY,conn)
df_AgeWork.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize = (10,7), color='darkblue')
plt.title('Scatter plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
group by age
order by age
"""
df_AgeWorkCod= pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.scatter(df_AgeWorkCod['WorkWeekHrs'], df_AgeWorkCod['CodeRevHrs'], s=df_AgeWorkCod['Age']*10, alpha=0.5)

# Adding labels and title
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours (Age as bubble size)')

# Display plot
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import matplotlib.pyplot as plt
import pandas as pd

QUERY = """
SELECT DatabaseDesireNextYear, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

df2 = pd.read_sql_query(QUERY, conn)
df2.set_index('DatabaseDesireNextYear', inplace=True)
color_list = ['blue', 'green', 'red', 'purple', 'yellow']
labels = df2.index[:5]

plt.figure(figsize=(10, 6))  # Set the figure size
plt.pie(df2['Count'], labels=labels, colors=color_list, autopct='%1.1f%%', startangle=90, shadow=True, pctdistance=0.85)
plt.title('This goes up top')  # Add the title
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(labels, loc='upper left')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

df_wrkwk = pd.read_sql_query(QUERY, conn)
df_wrkwk.head()

df_graph = df_wrkwk.groupby("Age").median()
df_graph.plot(kind='bar', stacked=True)
plt.title('Median workweek hours and CodeRevHrs in the 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df_Convert = pd.read_sql_query(QUERY, conn)
df_Convert.head()

df_gr = df_Convert.groupby("ConvertedComp").median()
df_gr.plot(kind='line')
plt.title('Median ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Age')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""

df_Mainbr = pd.read_sql_query(QUERY, conn)
df_Mainbr.head()

mainbranch_counts = df_Mainbr.value_counts()

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue')

# Adding labels and title
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Horizontal Bar Chart of Main Branch')

# Display plot
plt.grid(True)
plt.show()
```
